```python
dataset = df
```

```python
del dataset['satisfaction_encode']
```

```python
len(dataset)
```

```python
dataset = dataset.dropna()
len(dataset)
```

```python
dataset['Gender'] = encoder.fit_transform(dataset['Gender'])
dataset['Type of Travel'] = encoder.fit_transform(dataset['Type of Travel'])
dataset['Class'] = encoder.fit_transform(dataset['Class'])
dataset['Customer Type'] = encoder.fit_transform(dataset['Customer Type'])
dataset['satisfaction'] = encoder.fit_transform(dataset['satisfaction'])
```

```python
X = dataset.loc[:, dataset.columns != 'satisfaction'].values
y = dataset['satisfaction'].values
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(X,y, random_state=112, test_size=0.3)
```

```python
x_train
```

```python
y_train
```

```python
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier()
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred))
```
